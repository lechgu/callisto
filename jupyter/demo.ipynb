```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)
time = np.arange(0, 100, 1)
values = np.random.randn(100).cumsum()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(time, values, label='Time Series')
plt.xlabel('Time')
plt.ylabel('Value')
plt.title('Random Time Series Data')
plt.legend()
plt.grid(True)
plt.show()
```
